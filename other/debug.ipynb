```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
csv = pd.read_csv('log/log_bug.csv')

fig, ax1 = plt.subplots(figsize=(10,5),dpi=200)
plt.grid(True)
ax1.plot(csv.time.values)
ax2 = ax1.twinx()

x = csv.combination.values
for i in range(len(x)):
    #print(type(x[i]))
    x[i] = x[i].count(' ')
    
ax2.plot(x, color='red')

plt.show()
```
